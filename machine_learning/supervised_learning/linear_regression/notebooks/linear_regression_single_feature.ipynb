```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = [
  [650, 70],
  [800, 85],
  [900, 78],
  [1100, 112],
  [1200, 120],
  [1400, 135],
  [1500, 160],
  [1700, 170],
  [1850, 175],
  [2000, 210]
]

xs = [k[0] for k in data]
ys = [k[1] for k in data]
m = len(data)
```

```python
# plt.plot(xs,ys)
plt.scatter(xs,ys)
```

```python
# Step 1: Initialize parameters
w = 0.0  # slope
b = 0.0  # intercept
learning_rate = 0.000001  # small learning rate due to large x values
epochs = 1000  # number of iterations
tolerance = 1e-6
prev_cost = float('inf')

cost_history = []

# Step 2: Gradient Descent Loop
for epoch in range(epochs):
    dw = 0.0
    db = 0.0
    
    for i in range(m):
        x = xs[i]
        y = ys[i]
        y_pred = w * x + b
        error = y_pred - y
        dw += error * x
        db += error

    dw /= m
    db /= m

    w -= learning_rate * dw
    b -= learning_rate * db

    if epoch % 100 == 0:
        # Optional: print progress
        cost = sum((w * xs[i] + b - ys[i]) ** 2 for i in range(m)) / m
        cost_history.append(cost)
        if (abs(cost - prev_cost) < tolerance):
            print(f"Stopped at epoch {epoch}")
            break
        prev_cost = cost
        print(f"Epoch {epoch}: w = {w:.4f}, b = {b:.4f}, cost = {cost:.4f}")

# Final model
print("\nTrained model:")
print(f"w = {w:.4f}, b = {b:.4f}")
```

```python
# Plot original data
plt.scatter(xs, ys, color='blue', label='Data Points')

# Plot regression line
predicted_ys = [w * x + b for x in xs]
plt.plot(xs, predicted_ys, color='red', label='Regression Line')

plt.xlabel('Size (sqft)')
plt.ylabel('Price ($1000s)')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
```
